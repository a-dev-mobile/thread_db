```python
def read_csv(patch):
    df = pd.read_csv(patch, sep=',', index_col=False, dtype='unicode')
    df = df.astype(object).replace(np.nan, '')
    
    return df
```

```python
def leght_int(digit):
    i = 0
    for c in str(digit):
        i = i+1 
    return i 


def getLexicographicallyNumber(max_value,input_value):
    leght_input_value = 0
    leght_max_value = leght_int(max_value)
    leght_input_value=leght_int(input_value)
    result_leght = leght_max_value-leght_input_value
    result = ("0"*result_leght)+str(input_value)
    return result
```

```python

patch = '../csv/m Thread - 1_gost_8724_2002_diam_pitch.csv'
df_csv = read_csv(patch)
# ******************************
path = '/thread/metric/ru/standart/gost_8724_2002/diam_pitch'
database_ref = db.reference(path)
database_ref.delete()
# ******************************
# id,range_1,range_2,range_3,coarse_pitch,pitchs
d={}
class_id = {}
i = 0
df = df_csv

sum_row = df.shape[0]

for index, row in df.iterrows():
    print('----------------------------')
    i = i + 1
    print('{0} = {1}'.format(i, sum_row))
    keyJson = getLexicographicallyNumber(1234,i)
     
    d['id'] =row['id']
    d['range_1'] =row['range_1']
    d['range_2'] =row['range_2']
    d['range_3'] =row['range_3']
    d['coarse_pitch'] =row['coarse_pitch']
    d['pitchs'] =row['pitchs'].split(',')
  
    database_ref = db.reference(path+'/'+keyJson+'/')
    database_ref.set(d)
    d.clear()
```
